```python
# importing necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from pprint import pprint

```

```python
# fetching API data
def format_value(txt):
    lower, highest = txt.split(' - ')
    return (int(lower) + int(highest)) / 2 
```

```python
# fetching API data
 
url = 'https://api.thecatapi.com/v1/breeds'
response = requests.get(url)
data = response.json()
cats = []
for cat in data:
    average_weight = format_value(cat['weight']['metric'])
    average_life_span = format_value(cat['life_span'])
    reference_image_id = cat.get('reference_image_id')
    image_url =  f'https://cdn2.thecatapi.com/images/{reference_image_id}.jpg' if reference_image_id else ''
    dct = {
        'id':cat['id'],
        'name': cat['name'],
        'origin': cat['origin'],
        'description':cat['description'],
         'weight':average_weight,
        'life_span': average_life_span,
        'temperament':cat['temperament'],
         'image_url': image_url
    }
    cats.append(dct)
pprint(cats[:3])
```

```python
sorted_origin = df.groupby('Origin')['Origin'].count().sort_values()

countries = sorted_origin.index
values = sorted_origin.values
plt.bar(countries, values)
plt.xlabel('Countries')
plt.ylabel('Cats breeds')
plt.title('Cats breeds per country')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.savefig('cats-breed.png')
plt.show()
```

```python



fig, ax = plt.subplots()
ax.pie(values, labels = countries, autopct= '%1.0f', rotatelabels=True)
plt.show()
```
