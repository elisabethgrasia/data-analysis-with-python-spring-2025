```python
# import all the necessary packages
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import datetime
from pprint import pprint # pretty print

### The following library is needed to fetch data from the Yahoo Finance service
### For fetching data from Yahoo Finance
import yfinance as yf

### Style affects on layout of graphics
### Available styles can be listed as plt.style.available
sns.set()

```

```python
### Open the data in the dataframes named correspondingly.
## Downloading Nokia's and Ericson's historical stock prices

nokia = yf.download('NOKIA.HE', start = '2000-1-1')
ericson = yf.download('ERIBR.HE', start = '2000-1-1')

## In order to make the index looks nicer, it is good to modify the index using the below code
nokia.index = nokia.index.tz_localize(None)
ericson.index = ericson.index.tz_localize(None)
```

```python
### Visualizing Nokia's and Ericson's historical prices, using the  daily closing prices.
### Plot line graph for the two stocks

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
nokia['Close'].plot(ax=axs[0], color='#005AFF')
ericson['Close'].plot(ax=axs[1],  color='#002561')

### Add title and vertical axis-titles

axs[0].set_title('Nokia historical stock prices since 2000')
axs[1].set_title('Ericson historical stock prices since 2000')

axs[0].set_ylabel('Euros', loc='center')
axs[1].set_ylabel('Euros', loc='center')
plt.legend()
plt.show()
```
